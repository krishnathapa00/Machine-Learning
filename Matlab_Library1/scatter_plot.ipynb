```python
# Scatter Plot:in simple terms, a scatter plot is a type of graph that uses dots to represent the relationship between two variables.   
#  Each dot on the graph represents a data point, and the position of the dot on the x and y axes indicates the values of the two variables being compared. Scatter plots are commonly used to visualize correlations or patterns in data, making it easier to identify trends or outliers.
# Line Plot: A line plot is a type of graph that uses a continuous line to connect data points. It is commonly used to show trends over time or to compare multiple sets of data. Each point on the line represents a specific value, and the line helps to visualize the overall pattern or trend in the data. Line plots are useful for identifying changes, trends, and relationships between variables.
# Bar Plot: A bar plot, also known as a bar chart, is a type of graph that uses rectangular bars to represent data. Each bar's length is proportional to the value it represents. Bar plots are commonly used to compare different categories or groups of data. The bars can be displayed vertically or horizontally, and they are often used to show the frequency, count, or other measures of   data across different categories. Bar plots are effective for visualizing differences between groups and identifying trends in categorical data.        

# syntax: plot(x, y, 'format')
#1. x and y: These are the data points that you want to plot. x represents the values on the x-axis, and y represents the values on the y-axis. They can be lists, arrays, or any other iterable data structure.  
#2. format: This is an optional argument that specifies the style of the plot. It can include various characters to indicate the type of plot, color, and marker style. For example, 'ro' would create a red circle marker, while 'b-' would create a blue line. The format string can be customized to achieve different visual effects in the plot.


# Example: plt.scatter(x, y, color='red', marker='o') creates a scatter plot with red circular markers.
# plt.plot(x, y, color='blue', linestyle='--') creates a line plot with blue dashed lines.

import matplotlib.pyplot as plt
hours_studies=[1, 2, 3, 4, 5,6,7,8]
scores=[50, 60, 70, 80, 90,95,98,99]
plt.scatter(hours_studies, scores, color='red', marker='^')
plt.title('Hours of Study vs Scores')
plt.xlabel('Hours of Study')
plt.ylabel('Scores')
plt.grid(True)
plt.show()  

# uses scatter plots:
#1. Correlation Analysis: Scatter plots are commonly used to analyze the correlation between two variables. By plotting the data points, you can visually assess whether there is a positive, negative, or no correlation between the variables. This helps in understanding the relationship and strength of association between them.


```

```python
 # comparision of two classes:

import matplotlib.pyplot as plt
plt.scatter([1,2,3],[50,60,70], color='red', marker='^') # Class 1
plt.scatter([1,2,3],[55,65,75], color='blue', marker='o') # Class 2
plt.legend(['Class 1', 'Class 2'])
plt.title('Comparison of Two Classes')
plt.xlabel('Hours of Study')
plt.ylabel('Scores')
plt.grid(True)
plt.show()  

```
