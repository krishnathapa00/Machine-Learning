```python
# Saving Figures
# Matplotlib provides the savefig() function to save a figure to a file. You can specify the filename and format, and Matplotlib will handle the rest. Here's how you can save a figure:

# syntax: plt.savefig('filename.format', dpi=300, bbox_inches='tight') # dpi is the resolution of the saved figure, and bbox_inches='tight' ensures that the saved figure includes all elements without extra whitespace.
import matplotlib.pyplot as plt

# Create a simple plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 11]
plt.plot(x, y, color='blue', marker='o')
plt.title('Sample Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')    
# Save the figure
plt.savefig('folder_name/sample_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
