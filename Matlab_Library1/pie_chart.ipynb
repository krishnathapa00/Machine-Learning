```python
# Pie Chart show the proportion of each category in a dataset. They are useful for visualizing the distribution of categorical data and comparing the relative sizes of different categories. In Python, you can create pie charts using libraries such as Matplotlib or Seaborn. Here are some common functions used for creating and customizing pie charts:
    #1. pie(): This function is used to create a pie chart. It takes a list of values as input and generates a pie chart based on those values.
    #2. explode(): This function is used to "explode" or separate a slice of the pie chart from the rest. It takes a list of values as input, where each value corresponds to the fraction of the radius by which to offset each slice.
    #3. autopct(): This function is used to display the percentage value of each slice on the pie chart. It takes a string format as input, which specifies how the percentage should be displayed (e.g., '%1.1f%%' for one decimal place).
    #4. startangle(): This function is used to specify the starting angle of the pie chart. It takes a value in degrees as input, which determines the rotation of the pie chart.
    #5. labels(): This function is used to add labels to each slice of the pie chart. It takes a list of strings as input, where each string corresponds to the label for each slice.
    #6. colors(): This function is used to specify the colors of the slices in the pie chart. It takes a list of color values as input, where each value corresponds to the color of each slice.        
    #7. shadow(): This function is used to add a shadow effect to the pie chart. It takes a boolean value as input, which determines whether to display the shadow or not.
    #8. radius(): This function is used to specify the radius of the pie chart. It takes a value as input, which determines the size of the pie chart.
    #9. title(): This function is used to add a title to the pie chart. It takes a string as
    #10. input and displays it as the title of the chart.  

# Syntax: plt.pie(values, explode=explode_values, autopct='%1.1f%%', startangle=90, labels=labels, colors=colors, shadow=True, radius=1.0)
    # Example:1
import matplotlib.pyplot as plt
category = ['Category A', 'Category B', 'Category C']
number = [30, 45, 25]
explode_values = [0.05, 0, 0]  # Explode the first slice
plt.pie(number, explode=explode_values, autopct='%1.1f%%', startangle=90, labels=category, colors=['red', 'blue', 'green'], shadow=True, radius=1.0)
plt.title('Pie Chart Example')
plt.show()
```
