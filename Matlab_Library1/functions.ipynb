```python
# Matplot-is a python library used for creating static, animated, and interactive visualizations. 
# It provides a wide range of plotting functions and customization options to create various types of plots, 
# such as line plots, bar charts, scatter plots, histograms, and more. Matplotlib is widely used in 
# data analysis, scientific research, and machine learning for visualizing data and presenting results effectively.

# functions in matplotlib:
#1. plot(): This function is used to create a line plot. It takes x and y values as input and plots them on a graph.
#2. scatter(): This function is used to create a scatter plot. It takes x and y values as input and plots them as individual points on a graph.
#3. bar(): This function is used to create a bar chart. It takes x and y values as input and plots them as bars on a graph.
#4. hist(): This function is used to create a histogram. It takes a list of values as input and plots a histogram to show the distribution of the data.
#5. pie(): This function is used to create a pie chart. It takes a list of values and corresponding labels as input and plots a pie chart to show the proportion of each category.
#6. title(): This function is used to add a title to the plot. It takes a string as input and sets it as the title of the graph.
#7. xlabel(): This function is used to add a label to the x-axis. It takes a string as input and sets it as the label for the x-axis.
#8. ylabel(): This function is used to add a label to the y-axis. It takes a string as input and sets it as the label for the y-axis.
#9. show(): This function is used to display the plot. It renders the graph on the screen and allows you to visualize the data. 
#10. savefig(): This function is used to save the plot as an image file. It takes a filename as input and saves the current plot to that file.  
#11. legend(): This function is used to add a legend to the plot. It takes a list of labels as input and displays a legend that identifies each plot element.   
#12. grid(): This function is used to add a grid to the plot. It takes a boolean value as input and displays a grid on the graph if set to True.    
#13. subplot(): This function is used to create multiple plots in a single figure. It takes the number of rows, columns, and the index of the current plot as input and creates a subplot accordingly.          
#14. xlim() and ylim(): These functions are used to set the limits of the x and y axes, respectively. They take two values as input (the lower and upper limits) and adjust the axes accordingly.   
#15. xticks() and yticks(): These functions are used to set the tick marks on the x and y axes, respectively. They take a list of values as input and set the tick marks at those positions on the axes.    
#16. annotate(): This function is used to add annotations to the plot. It takes a string as input and adds it as an annotation at a specified location on the graph.    
#17. color(): This function is used to set the color of the plot elements. It takes a string or a list of colors as input and applies it to the corresponding plot elements.    












# First example: Line plot
import matplotlib.pyplot as plt
x=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'] # x-axis values
y=[10, 20, 15, 25, 30] # y-axis values
plt.plot(x, y) # create a line plot
plt.title('Bakery Sales') # add title

plt.xlabel('Days of the Week') # add x-axis label
plt.ylabel('Number of Sales') # add y-axis label
plt.show() # display the plot.
```

```python
import matplotlib.pyplot as plt
month=[1,2,3,4,5] # x-axis values
salses=[1000, 1200, 800, 2500, 1200] # y-axis values
plt.plot(month, salses, marker='o', linestyle='--', color='blue', linewidth=2, label=' 2025 Sales Data') # create a line plot with markers and custom style
plt.title('Monthly Sales Data') # add title
plt.xlabel('Months') # add x-axis label
plt.ylabel('Sales Per Month') # add y-axis label
plt.legend(loc='upper left', fontsize=10) # add legend to the plot-it shows the label defined in the plot function
plt.grid(color='gray', linestyle='--', linewidth=0.5) # add grid to the plot
#plt.xlim(0, 6) # set x-axis limits
#plt.ylim(0, 3000) # set y-axis limits
plt.xticks([1,2,3,4,5], ['Jan','Feb','Mar','Apr','May']) # set x-axis ticks to the month values with labels
plt.show() # display the plot.
```
