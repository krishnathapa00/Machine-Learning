```python
# Plotting Histogram of a Dataset.
# syntx: hist(data, bins)
# data: This is the dataset for which you want to create the histogram. It can be a list or an array of numerical values.
# bins: This is the number of bins you want to use for the histogram. It can be an integer or a list of bin edges. If you provide an integer, it will automatically create that many bins between the minimum and maximum values of the dataset. If you provide a list of bin edges, it will use those specific edges to create the bins for the histogram. 

#plt.hist(data, bins=number,colors='colorname',edgecolor='black')  # Example usage
import matplotlib.pyplot as plt
import numpy as np
# Sample data
#data = np.random.randn(1000)  # Generate 1000 random numbers from a normal distribution
scores= [85, 90, 78, 92, 88, 95, 80, 82, 91, 89]  # Example dataset of scores
# Create histogram
plt.hist(scores, bins=5, color='blue', edgecolor='black')
# Add title and labels
plt.title('Histogram of Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
# Show the plot
plt.show()
```
