```python
# Ploting a bar chart using matplotlib: plt.bar(x, height, width=0.8, color='blue', edgecolor='black') where x is the position of the bars on the x-axis, height is the height of the bars, width is the width of the bars (default is 0.8), color is the color of the bars (default is blue), and edgecolor is the color of the edges of the bars (default is black).
# bar chart is used representing categorial data.
import matplotlib.pyplot as plt
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [100, 150, 80, 120]
plt.bar(products, sales, color='orange', edgecolor='black')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales of Products in 2025')
plt.legend(['Sales 2025'])
# display plot horizontally using plt.barh() instead of plt.bar()
plt.show()  
```
